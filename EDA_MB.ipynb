# Missing values and Imputation

```python
import seaborn as sns
cols=df_cleaned3.columns
colours = ['#000099','#ffff00']
sns.heatmap(df_cleaned3[cols].isnull(),
           cmap=sns.color_palette(colours))
```

```python
df_cleaned3[df_cleaned3.isnull().any(axis=1)]
```

```python
df_cleaned3[df_cleaned3.isnull().any(axis=1)].shape
```

```python
df_cleaned3.isnull().sum()
```

```python
# Mean Imputation
mean= df_cleaned3['Ozone'].mean()
```

```python
df_cleaned3
```

```python
df_cleaned3['Ozone']= df_cleaned3['Ozone'].fillna(mean)
```

```python
df_cleaned3
```

```python
df_cleaned3.info()
```

```python
# Missing value imputation for categorical value
#Get the object column
obj_columns=df_cleaned3[['Weather']]
```

```python
obj_columns
```

```python
obj_columns.isnull().sum()
```

```python
# Missing value imputation for categorical value
obj_columns=obj_columns.fillna(obj_columns.mode().iloc[0])
```

```python
obj_columns.mode()
```

```python
obj_columns.isnull().sum()
```

```python
df_cleaned3=df_cleaned3.drop('Weather',axis=1)
```

```python
df_cleaned3
```

```python
df_cleaned3.shape
```

```python
obj_columns.shape
```

```python
# Join the data set with imputed object dataset
df_cleaned4=pd.concat([df_cleaned3,obj_columns],axis=1)
```

```python
df_cleaned4
```

# Feature Scaling

```python
# Normalization of the data
from numpy import set_printoptions
from sklearn.preprocessing import MinMaxScaler
```

```python
array=df_cleaned4.values
array
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
rescaledX= scaler.fit_transform(array[:,0:5])
                    
#summarize transformed data
set_printoptions (precision=2)
print(rescaledX[0:5,:])                    
                    
```

```python
#standardize data (0 mean, 1 stdev)
from sklearn.preprocessing import StandardScaler
```

```python
array=df_cleaned4.values
scaler= StandardScaler().fit(array)
rescaledX= scaler.transform(array)

#Summarize transformed data
set_printoptions(precision=2)
print(rescaledX[0:5,:])
```
